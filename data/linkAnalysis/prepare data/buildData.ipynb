```python
# imports
import pandas as pd
import numpy as np
```

```python
## -------------------------------------- Reading and basic cleaning -------------------------------------- ##
```

```python
# Read the dataset
chicago_crimes = pd.read_csv('./chicago_crimes_clean.csv')
```

```python
chicago_crimes = pd.DataFrame(chicago_crimes)
chicago_crimes
```

```python
chicago_crimes['Date'] = pd.to_datetime(chicago_crimes['Date'])
```

```python
chicago_crimes['Weekday'] = chicago_crimes['Date'].dt.day_name().str[:3]
chicago_crimes['Month'] = chicago_crimes['Date'].dt.month_name().str[:3]
```

```python
def fWeekMonth(x):
    if (x>=0) & (x<8):
        wMonth = 'First'
    elif (x>=8) & (x<15):
        wMonth = 'Second'
    elif (x>=15) & (x<22):
        wMonth = 'Third'
    else:
        wMonth = 'Fourth'
    return wMonth

chicago_crimes['WeekMonth'] = chicago_crimes.Date.dt.day.map(fWeekMonth) 
```

```python
# Fix misspelling
chicago_crimes['Primary Type'] = np.where(chicago_crimes['Primary Type'] == 'CRIM SEXUAL ASSAULT', 'CRIMINAL SEXUAL ASSAULT', chicago_crimes['Primary Type'])
```

```python
chicago_crimes['Primary Type'] = chicago_crimes['Primary Type'].str.replace(' ', '_')
```
